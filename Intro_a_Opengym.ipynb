## Operaciones básicas en entornos

Leemos el entorno que queremos. Hay muchos entornos posibles, aquí seleccionamos el 'Taxi-v2'

```python
env = gym.make("Taxi-v3") 
```

Para cerrar el entorno entorno

```python
env.close()
```

Reseteamos el entorno

```python
env.reset()
```

Ploteamos el entorno

```python
env.render()
```

El entorno es discreto, tiene un número de posibles estados finito. En este caso 500

```python
env.observation_space
```

Ponemos un estado en particular (114)

```python
env.env.s = 114
env.render()
```

Cuantas posibles acciones tenemos

```python
env.action_space.n
```

Loop random movement

```python
env.env.s = 1
state = env.reset()
counter = 0
reward = -1
while reward < 0 & counter<9900:
    aa = env.step(env.action_space.sample())
    
    print(aa)
    reward = aa[1]
    env.render()
    counter += 1
    print(counter)
    #if aa[2] == True:
    #    state = env.reset()
print(counter)
```
